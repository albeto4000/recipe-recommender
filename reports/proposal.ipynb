## Abstract

&emsp; Home cooking is a necessary yet surprisngly high-friction task. Home cooks often struggle with decision fatigue, meaning that despite the countless number of recipes, guides, and cookbooks, home cooks still find themselves asking "What's for dinner?" In addition to the cook's tastes, the home cook also has to consider what ingredients are available to them, what ingredients need to be acquired, is this a recipe that they can feasibly accomplish after work, and is this a recipe that fits within their nutrition plan. The proposed project aims to assist home cooks in making these decisions through both content- and knowledge-based recommendations. Recommendations will be filtered by user's preferences in tastes, what ingredients they have on hand, difficulty, and nutrition. Recommendations will then be ranked according to what the user has enjoyed in the past, by difficulty, and by popularity. Additionally, users will have the option to add ingredients to a shopping cart which can used as a shopping list or as a budgeting tool. Through the proposed project, we hope to reduce the barrier to home cooking by creating a recommendation system that assists with the decision making and planning process of cooking.

## Project Goals

&emsp; The goal of the project is to create a recipe recommendation system that forms both content- and knowledge-based recommendations. Content-based recommendations will be formed by utilizing the ingredient list, difficulty, cuisine, and the nutritional facts. Knowledge-based recommendations will be formed by using the user's taste preferences through what foods and recipes they enjoy, the user's current set of available ingredients, the user's nutrition plan, and the user's skill level. 

&emsp; In addition to forming recommendations, we plan to create a shopping cart that users can utilize as a shopping list or as a budgeting tool.

## Project Plan


&emsp; The major hurdles for the project are to (1) process the data into a suitable form for forming recommendations, (2) create recommendation models, and (3) create a user interface. Data required for this project is tabular and comes from Food.com (formerly GeniusKitchen) hosted on Kaggle.

&emsp; Data processing will cleaning, standardizing, and feature engineering. Processing can be completed in a few days but will likely require revisits throughout the development process.

&emsp; The recommendation phase will take up the majority of the development time. We envision developing multiple recommendation systems with each prioritizing a specific use case such as ease (under 30 minutes), familiarity, popularity, or discovery. We will also need to consider user preferences in these recommendations and their rankings.

&emsp; We plan to create an interactive user interface that can be hosted on a web server. Users will be able to add recipes to a shopping cart.

## Data Sources and EDA

&emsp; The dataset we intend to build our recommender system on - Food.com Recipes and Interactions - was sourced from Kaggle. It consists of over a million ratings of 231,637 recipes by 226,570 Food.com users. Each recipe is described by its ID, name, author ID, date submitted, ingredients, instructions (steps), tags, nutrition, and description. The nutrition column is a list containing each recipe’s calories, total fat, sugar, sodium , protein , and saturated fat; aside from calories, all values are measured as % of daily value (PDV). Other numeric columns describe the number of ingredients and steps. From this information, we intend to derive each recipe’s region of origin (cuisine) and estimate its difficulty. Rating data consists of recipe IDs, user IDs, ratings (out of five stars), and review text. 
The density of the ratings matrix is 2.15 * 10-5. Approximately 73% of all users have only submitted a single rating, and 70% of all recipes have 3 ratings or fewer. Due to this sparsity, we expect that collaborative filtering techniques will be difficult to implement using this dataset. Therefore, we will primarily be exploring content-based filtering and knowledge-based recommendation.
