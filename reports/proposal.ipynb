### Data Sources and EDA

&emsp; The dataset we intend to build our recommender system on - Food.com Recipes and Interactions - was sourced from Kaggle. It consists of over a million ratings of 231,637 recipes by 226,570 Food.com users. Each recipe is described by its ID, name, author ID, date submitted, ingredients, instructions (steps), tags, nutrition, and description. The nutrition column is a list containing each recipe’s calories, total fat, sugar, sodium , protein , and saturated fat; aside from calories, all values are measured as % of daily value (PDV). Other numeric columns describe the number of ingredients and steps. From this information, we intend to derive each recipe’s region of origin (cuisine) and estimate its difficulty. Rating data consists of recipe IDs, user IDs, ratings (out of five stars), and review text. 
The density of the ratings matrix is 2.15 * 10-5. Approximately 73% of all users have only submitted a single rating, and 70% of all recipes have 3 ratings or fewer. Due to this sparsity, we expect that collaborative filtering techniques will be difficult to implement using this dataset. Therefore, we will primarily be exploring content-based filtering and knowledge-based recommendation.

### Expected Complications

1. **Data imbalance** \
&emsp; We will have to find a way to deal with the explicit data bias as users are most likely going to leave reviews on recipes that they really like (5) or don't like (0), making it difficult for models to predict negative ratings. 
2. **Sparse ratings matrix** \
&emsp; A primary challenge will be the high degree of data sparsity inherent in the dataset. With a large ratio of recipes to active reviewers, the user-item interaction matrix will be sparsely populated, potentially leading to 'cold start' issues for new recipes and users with niche tastes.
3. **Feature Engineering from Unstructured Data** \
&emsp; Part of our plan will be to create additional categories to help further filter and differentiate recipes from one another. One issue will be that some techniques used within each recipe may be the same but can be worded differently, so our task will be to create a level of uniformity between these similar steps. Another issue would be pertaining to the ingredients as many ingredients have various names so we will have to decide on what will be the best way to rename ingredients to ensure that they match.
4. **Calculation Complexity** \
&emsp; We are dealing with millions of recipes and hundreds of thousands of users, so calculating similarity matrices may be computationally expensive, thus looking into possible dimensionality reduction techniques that do not make any sacrifices to accuracy may be a solution to look into.
5. **Validation Hurdles** \
&emsp; As this project will be conducted, it will be difficult to truly gauge how successful our recommender will be without A/B testing, so we will need to make use of offline evaluation metrics such as RMSE (Root Mean Squared Error) and MAE (Mean Absolute Error) to validate our systems performance.
